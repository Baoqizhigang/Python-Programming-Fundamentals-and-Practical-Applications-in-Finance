```python
# Seaborn is a visualization library that sits on top of matplotlib
# Seaborn offers enhanced features compared to matplotlib
# https://seaborn.pydata.org/examples/index.html
# import libraries

import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical naalysis
import matplotlib.pyplot as plt # Import matplotlib for data visualisation\
import seaborn as sns # Statistical data visualization 
```

```python
# np.C_ class object translates slice objects to concatenation along the second axis.
# np.array: Creates an array from the provided list. x1 and x2 are simple 1D NumPy arrays.
x1 = np.array([1,2,3])
x1.shape
```

```python
x2 = np.array([4,5,6])
x2.shape
```

```python
# np.c_: A shorthand for np.column_stack which combines arrays along the second axis (column-wise), resulting in z being a 2D array combining x1 and x2.
z = np.c_[x1,x2]
print(z)
print(z.shape)
```

```python
# Import Cancer data drom the Sklearn library
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer
```

```python
# Create a dataFrame named df_cancer with input/output data
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'],['target']))
```

```python
# Check out the head of the dataframe
df_cancer.head()
```

```python
# Check out the tail of the dataframe
df_cancer.tail()
```

```python
# Plot scatter plot between mean area and mean smoothness
# hue adds color coding based on the 'target' column,
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target', data = df_cancer)
```

```python
# Let's print out countplot to know how many samples belong to class #0 and #1
sns.countplot(df_cancer['target'], label = 'Count')
```

```python
sns.scatterplot(x = 'mean radius', y = 'mean area', hue = 'target', data = df_cancer)
```

```python
# Plot the pairplot
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'])
```

```python
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
plt.figure(figsize = (20, 10))
sns.heatmap(df_cancer.corr(), annot=True)
```

```python
# plot the distplot 
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. 

sns.displot(df_cancer['mean radius'], bins = 25, color = 'blue')
```

```python
# Plot two separate distplot for each target class #0 and target class #1
class_0_df = df_cancer[ df_cancer['target'] == 0]
class_1_df = df_cancer[ df_cancer['target'] == 1]
```

```python

plt.figure(figsize = (10, 7))
sns.histplot(class_0_df['mean radius'], bins = 50, color = 'blue', kde = True, stat = 'density', kde_kws = dict(cut = 3), alpha = 0.4, edgecolor=(1, 1, 1, .4))
sns.histplot(class_0_df['mean radius'], bins = 50, color = 'red', kde = True, stat = 'density', kde_kws = dict(cut = 3), alpha = 0.4, edgecolor=(1, 1, 1, .4))
sns.rugplot(class_0_df, x="mean radius")
plt.grid()
```
